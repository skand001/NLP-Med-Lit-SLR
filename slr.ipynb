```python
print(f"Average ROUGE-1: {df['ROUGE-1'].mean():.3f}")
print(f"Average ROUGE-2: {df['ROUGE-2'].mean():.3f}")
print(f"Average ROUGE-L: {df['ROUGE-L'].mean():.3f}")
print(f"Average Accuracy: {df['Accuracy'].mean():.3f}")
```

```python
grouped_by_year = df.groupby('Year').mean()
print(grouped_by_year)
```

```python
filtered_df = df[(df['ROUGE-1'] > 50) & (df['ROUGE-2'] > 20)]
print(filtered_df)
```

```python
# Aggregation: Group by 'Year' and calculate the mean of 'ROUGE-1', 'ROUGE-2', 'ROUGE-L', and 'Accuracy'
grouped_by_year = df.groupby('Year')[['ROUGE-1', 'ROUGE-2', 'ROUGE-L', 'Accuracy']].mean()

# Print the aggregated results
print("Aggregated Results:")
print(grouped_by_year)

# Visualization: Bar plot of average ROUGE scores by year
fig, ax = plt.subplots(figsize=(10, 6))
grouped_by_year[['ROUGE-1', 'ROUGE-2', 'ROUGE-L']].plot(kind='bar', ax=ax)
ax.set_xlabel('Year')
ax.set_ylabel('Average Score')
ax.set_title('Average ROUGE Scores by Year', fontweight='bold', fontsize=14)
ax.legend(['ROUGE-1', 'ROUGE-2', 'ROUGE-L'])

# Adjust the spacing between subplots
plt.tight_layout()
plt.grid(True)
# Display the plot
plt.show()
```

```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt



# Concatenate all the text from the 'Method' column into a single string
text = ' '.join(df['Method'].astype(str))

# Create a WordCloud object
wordcloud = WordCloud(width=800, height=400, background_color='black', colormap='viridis', max_words=50).generate(text)

# Create a figure and axes
fig, ax = plt.subplots(figsize=(12, 6))

# Display the word cloud
ax.imshow(wordcloud, interpolation='bilinear')
ax.axis('off')

# Set the title
ax.set_title('Word Cloud of Methods', fontweight='bold', fontsize=22)

# Adjust the layout
plt.tight_layout()

# Display the plot
plt.show()
```
